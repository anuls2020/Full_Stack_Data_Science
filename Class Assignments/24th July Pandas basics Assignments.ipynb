### 2. Do a bulk load for these two table for respective dataset

```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine

# Create Connection to MySQL Engine

engine = create_engine("mysql://root:root@localhost/DRESS_SHOP")

#Create dataframe from the the data source
df1 = pd.read_excel(r'Attribute DataSet.xlsx')
df2 = pd.read_excel(r'Dress Sales.xlsx')

#Create MySQL table using Pandas

df1.to_sql('attribute', con = engine, if_exists='replace', index=False)
df2.to_sql('dress', con = engine, if_exists= 'replace', index = False)
```

### 10. Try to find out a third highest most selling dress id

```python
select Dress_ID, SUM(`29/8/2013` + `31/8/2013` + `2013-02-09 00:00:00` + `2013-04-09 00:00:00` + `2013-06-09 00:00:00` +
        `2013-08-09 00:00:00` + `2013-10-09 00:00:00` + `2013-12-09 00:00:00` + `14/9/2013` + `16/9/2013` +
        `18/9/2013` + `20/9/2013` + `22/9/2013` + `24/9/2013` + `26/9/2013` + `28/9/2013` + `30/9/2013` +
        `2013-02-10 00:00:00` + `2013-04-10 00:00:00` + `2013-06-10 00:00:00` + `2010-08-10 00:00:00` +
        `2013-10-10 00:00:00` + `2013-12-10 00:00:00`) as TotalDressSell
from dress
group by Dress_ID
order by TotalDressSell desc
limit 1
offset 2
```
