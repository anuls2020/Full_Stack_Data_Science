### q9 : Write a code to send a mail to your friend 

```python
# Import the required libraries
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication

# Required email parameters
sender_email =  "xxxxyyxx@gmail.com"
receiver_email = "xxxx@gmail.com"
subject = "Testing Python code to send mail"
body = "This is a testing mail to check python mail sending code"

# Gmail SMTP server settings
smtp_server = "smtp.gmail.com"
smtp_port = 587  # TLS encryption port

# Gmail account credentials
gmail_username = "xxxx@gmail.com"
gmail_app_password = "***********"

# Create a multipart message and set the headers
message = MIMEMultipart()
message["From"] = sender_email
message["To"] = receiver_email
message["Subject"] = subject

# Attach the email body as plain text
message.attach(MIMEText(body, "plain"))

# Connect to SMTP server and send the email
try:
    with smtplib.SMTP(smtp_server, smtp_port) as server:
        # Start the TLS encryption
        server.starttls()
        
        # login to Gmail account
        server.login(gmail_username, gmail_app_password)
        
        # Send the email
        server.sendmail(sender_email, receiver_email, message.as_string())
    print("Email send successfully!..")
    
except Exception as e:
    print("An error occurred: ", str(e))
```

```python
# Importing required libraries
from datetime import datetime   #To set date and time
from playsound import playsound     #To play sound

def validate_time(alarm_time):
    if len(alarm_time) != 11:
        return "Invalid time format! Please try again..."
    else:
        if int(alarm_time[0:2]) > 12:
            return "Invalid HOUR format! Please try again..."
        elif int(alarm_time[3:5]) > 59:
            return "Invalid MINUTE format! Please try again..."
        elif int(alarm_time[6:8]) > 59:
            return "Invalid SECOND format! Please try again..."
        else:
            return "ok"

while True:
    alarm_time = input("Enter time in 'HH:MM:SS AM/PM' format: ")
    
    validate = validate_time(alarm_time.lower())
    if validate != "ok":
        print(validate)
    else:
        print(f"Setting alarm for {alarm_time}...")
        break

alarm_hour = alarm_time[0:2]
alarm_min = alarm_time[3:5]
alarm_sec = alarm_time[6:8]
alarm_period = alarm_time[9:].upper()

while True:
    now = datetime.now()

    current_hour = now.strftime("%I")
    current_min = now.strftime("%M")
    current_sec = now.strftime("%S")
    current_period = now.strftime("%p")

    if alarm_period == current_period:
        if alarm_hour == current_hour:
            if alarm_min == current_min:
                if alarm_sec == current_sec:
                    print("Wake Up!")
                    playsound(r"Path to the audio file")
                    break
```

### q20 : write a function by which you will be able to append two PDF files . 

```python
!pip install PyPDF2
```

```python
import PyPDF2

def merge_pdf(input_pdf1,input_pdf2,outputpdf):
    pdf_merger = PyPDF2.PdfMerger()
    
    #Open pdf files in read binary format
    with open(input_pdf1, 'rb') as file1:
        pdf_merger.append(file1)
        
    with open(input_pdf2, 'rb') as file2:
        pdf_merger.append(file2)
        
    # write merged pdf into output file
    with open(outputpdf, 'wb') as output_file:
        pdf_merger.write(output_file)
    
    print("PDFs appented successfully.")

input_pdf1  = '1.pdf'
input_pdf2 = '2.pdf'
outputpdf = 'output.pdf'
merge_pdf(input_pdf1,input_pdf2,outputpdf)
```

### q21 : write a function which can help you to filter only word file from a directory . 

```python
#Import required libraries
import os

# Function definition for word files
def filter_word_files(directory):
    word_files = []
    
    # list all files in the directory
    files = os.listdir(directory)
    
    # filter for files with .docx extension
    for file in files:
        if file.endswith('.docx'):
            word_files.append(file)
    return word_files
        
# Directory path
dir_path = r'Path/directory'

# Function call to extract the files.
word_files = filter_word_files(dir_path)

# print the directories
print("The word files in the directory are:")

for file in word_files:
    print(file)
```

### q22 : write a function which can read video file and play for you . 

```python
!pip install opencv-python
```

Downloaded the sample files form the site
https://file-examples.com/index.php/sample-video-files/

```python
# import libraries
import cv2

# video file path
video_path = r'Path to video file'

# video capture object
cap = cv2.VideoCapture(video_path)

# check the file opeed successfully
if not cap.isOpened():
    print("Error: Could not open the video file")
    exit()

# loop to read and display each frame
while True:
    # Read the frame form the video
    ret, frame = cap.read()
    
    # Check if the video have reached the end
    if not ret:
        break
        
    # Display the frame
    cv2.imshow('video player', frame)
    
    # exit the 'q' key is pressed
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break


# Release the video capture object and close the OpenCV Window

cap.release()
cv2.destroyAllWindows()
```

### q23 : write a function which will be able to shutdown your system.

```python
# Import libraries
import os
import platform

# define the shutdown function
def shutdown_system():
    # check system platform
    os_platform  = platform.system()
    
    if os_platform == 'Windows':
        os.system("shutdown /s /t 0")
        
    elif os_platform == "Linux" OR os_platform == "Darwin":
        os.system("shutdown -h now")
    
    else: 
        print("Unsupported OS")

# function call
#shutdown_system()
```
